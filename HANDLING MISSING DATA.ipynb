```python
import numpy as np
import pandas as pd
```

```python
# Read the csv file
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.info()
```

```python
#Lets Separate alphabet and number from "Number column" Pandas provide a function called as pd.to_numberic
df['no_number']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
# Now lets deal with alphabet 
df['no_categorical']=np.where(df['no_number'].isnull(),df['number'],np.nan)
```

```python
df
```

```python
df['Cabin']
```

```python
# As cabin contains Alphanumeric data, lets split it into number and alphabet
df['cabin_number']=df['Cabin'].str.extract('(\d+)')
# As cabin contain aphabetNumber. so we can extract alphabet by doing slice operation
df['cabin_catgorical']=df['Cabin'].str[0]
```

```python
df
```

```python
# Lets deal with Ticket number
df['Ticket_number']=df['Ticket'].apply(lambda a:a.split()[-1])
df['Ticket_number']=pd.to_numeric(df['Ticket_number'],errors='coerce',downcast='integer')
df['Ticket_catgorical']=df['Ticket'].apply(lambda s:s.split()[0])
df['Ticket_catgorical']=np.where(df['Ticket_catgorical'].str.isdigit(),np.nan,df['Ticket_catgorical'])
```

```python
df
```

```python
#DROP cabin,ticket,number,survived column
df=df.drop(['Cabin','Ticket','number'],axis=1)
```
